```python
import numpy as np

# 读取数据
data = np.loadtxt(r'C:\Users\WjtDg\Desktop\数据处理2新\train.txt', delimiter=',')

# 定义滑动窗口大小和步长
window_size = 20
step = 20 # 更改步长为 20

# 初始化滑动窗口特征列表和标签列表
window_features_list = []
window_labels_list = []

# 遍历数据并划分窗口
for i in range(0, len(data) - window_size + 1, step):  # 调整步长
    # 提取滑动窗口的特征和标签
    window_features = data[i:i+window_size, 0:-1]  # 提取特征列，排除时间列和类别标签列
    window_labels = data[i:i+window_size, -1]  # 提取滑动窗口内所有时间点的类别标签
    most_common_label = np.bincount(window_labels.astype(int)).argmax()  # 找到滑动窗口内出现次数最多的标签
    window_label = most_common_label  # 将出现次数最多的标签作为最终的标签
    # 将特征和标签分别添加到对应的列表中
    window_features_list.append(window_features)
    window_labels_list.append(window_label)

# 将特征列表和标签列表转换为numpy数组
window_features_array = np.array(window_features_list)
window_labels_array = np.array(window_labels_list)

# 打印特征和标签数组的形状
print("Window features array shape:", window_features_array.shape)
print("Window labels array shape:", window_labels_array.shape)

# 保存窗口特征和标签为.npy文件
np.save(r'C:\Users\WjtDg\Desktop\data0.npy', window_features_array)
np.save(r'C:\Users\WjtDg\Desktop\lable0.npy', window_labels_array)
```
