```python
import numpy as np

# è¯»å–æ•°æ®
data = np.loadtxt(r'C:\Users\WjtDg\Desktop\æ•°æ®å¤„ç†2æ–°\train.txt', delimiter=',')

# å®šä¹‰æ»‘åŠ¨çª—å£å¤§å°å’Œæ­¥é•¿
window_size = 20
step = 20 # æ›´æ”¹æ­¥é•¿ä¸º 20

# åˆå§‹åŒ–æ»‘åŠ¨çª—å£ç‰¹å¾åˆ—è¡¨å’Œæ ‡ç­¾åˆ—è¡¨
window_features_list = []
window_labels_list = []

# éå†æ•°æ®å¹¶åˆ’åˆ†çª—å£
for i in range(0, len(data) - window_size + 1, step):  # è°ƒæ•´æ­¥é•¿
    # æå–æ»‘åŠ¨çª—å£çš„ç‰¹å¾å’Œæ ‡ç­¾
    window_features = data[i:i+window_size, 0:-1]  # æå–ç‰¹å¾åˆ—ï¼Œæ’é™¤æ—¶é—´åˆ—å’Œç±»åˆ«æ ‡ç­¾åˆ—
    window_labels = data[i:i+window_size, -1]  # æå–æ»‘åŠ¨çª—å£å†…æ‰€æœ‰æ—¶é—´ç‚¹çš„ç±»åˆ«æ ‡ç­¾
    most_common_label = np.bincount(window_labels.astype(int)).argmax()  # æ‰¾åˆ°æ»‘åŠ¨çª—å£å†…å‡ºç°æ¬¡æ•°æœ€å¤šçš„æ ‡ç­¾
    window_label = most_common_label  # å°†å‡ºç°æ¬¡æ•°æœ€å¤šçš„æ ‡ç­¾ä½œä¸ºæœ€ç»ˆçš„æ ‡ç­¾
    # å°†ç‰¹å¾å’Œæ ‡ç­¾åˆ†åˆ«æ·»åŠ åˆ°å¯¹åº”çš„åˆ—è¡¨ä¸­
    window_features_list.append(window_features)
    window_labels_list.append(window_label)

# å°†ç‰¹å¾åˆ—è¡¨å’Œæ ‡ç­¾åˆ—è¡¨è½¬æ¢ä¸ºnumpyæ•°ç»„
window_features_array = np.array(window_features_list)
window_labels_array = np.array(window_labels_list)

# æ‰“å°ç‰¹å¾å’Œæ ‡ç­¾æ•°ç»„çš„å½¢çŠ¶
print("Window features array shape:", window_features_array.shape)
print("Window labels array shape:", window_labels_array.shape)

# ä¿å­˜çª—å£ç‰¹å¾å’Œæ ‡ç­¾ä¸º.npyæ–‡ä»¶
np.save(r'C:\Users\WjtDg\Desktop\data0.npy', window_features_array)
np.save(r'C:\Users\WjtDg\Desktop\lable0.npy', window_labels_array)
```
